# Define a function that takes a single string argument and returns the string backwards
- Call it with a string and print the result that is returned

```python
def get_backwards(string):
    return string[::-1]
print(get_backwards('Hello world'))
```

# Define a function that takes two strings as arguments and returns the longest one (or `None` if they are the same)
- Call it and print the result for three few different examples - one for each possibility

```python
def get_longer(string_1, string_2):
    if len(string_1) > len(string_2):
        return string_1
    elif len(string_1) < len(string_2):
        return string_2
    else:
        return None

print(get_longer('Hello World', 'Hello Everyone'))
print(get_longer('Hello Everyone', 'Hello World'))
print(get_longer('Hello World', 'Hello World'))
```

# Now write a function with three arguments - two strings to compare and a final string that contains either 'largest' or 'smallest' 
- Write the function so that the largest or smallest string is returned, depending on the value of the third parameter
- If both are the same length, return `None`
- If the third parameter is omitted, return the larger (default value)
- Run it a few times and print the result

```python
def get_largest_or_smallest(string_1='', string_2='', compare_string=''):
    if compare_string:
        string_1_diff = len(string_1) - len(compare_string)
        string_2_diff = len(string_2) - len(compare_string)
        if string_1_diff > string_2_diff:
            return string_1
        elif string_1_diff < string_2_diff:
            return string_2
        else:
            return None
    elif len(string_1) > len(string_2):
        return string_1
    elif len(string_1) < len(string_2):
        return string_2
    else:
        return None

print(get_largest_or_smallest('Hello World', 'Hello Everyone', 'Hello'))
print(get_largest_or_smallest('Hello World', 'Hello', 'Hello'))
print(get_largest_or_smallest('Hello World', 'Hello Everyone'))
```

# Now write a function that takes a list of strings and returns the longest string
- Call it with a few different examples

```python
my_strings = ['Hello', 'Hello World', 'Hello Everyone']
more_strings = ['Apple', 'Bananas', 'Oranges', 'Pineapple']

def get_longest_from_list(string_list):
    longest = ''
    for word in string_list:
        if len(word) > len(longest):
            longest = word
    return longest

print(get_longest_from_list(my_strings))
print(get_longest_from_list(more_strings))

```

# Write a function that takes a list of words to filter, and an integer that specifies a word length threshold
- The function returns a list of all the words that are longer than the specified length

```python
word_list = ['water', 'lantern', 'playing', 'price', 'toy', 'bag', 'us']

def get_long_words(list_words, length=0):
    long_words = []
    for word in list_words:
        if len(word) > length:
            long_words.append(word)
    return long_words

print(get_long_words(word_list,4))


```

# Write a function that returns true if the string passed to its single argument is a palindrome (same forwards as backwards)

```python
def check_palindrome(word_string):
    if word_string == word_string[::-1]:
        return True
    else:
        return False

print(check_palindrome('kayak'))
print(check_palindrome('hello'))
print(check_palindrome('peep'))
```
