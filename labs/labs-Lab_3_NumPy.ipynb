# Run the following two boxes of code to generate the data for this lab and plot it

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

rows=5
points=100
sins = np.zeros((rows,points))
for i in range(rows):
    for j in range(points):
        sins[i,j]=i*np.sin(j*(i+1)/20)
```

```python
for s in sins:
    plt.plot(s)
plt.show()
```

# Now use Numpy to perform the following tasks
## Extract the third series and plot it

```python
plt.plot(sins[2])
plt.show()
```

## Extract the 10th data point from each series and plot them as a series

```python
plt.plot(sins[0:5,10])
plt.show()
```

## Plot only those series that go above 2 at any point

```python
for s in sins:
    if (s>2).any():
        plt.plot(s)
plt.show()
```

## Plot the result of adding all the series together

```python
plt.plot(sins.sum(axis=0))
plt.show()
```

## Plot the result of adding the second, third and fourth rows together
- Use a range selection

```python
plt.plot(sins[1:4].sum(axis=0))
plt.show()
```

## Move the second line up by 2  and replot all the lines

```python
for i, s in enumerate(sins):
    if i == 1:
        plt.plot(s + 2)
    else:
        plt.plot(s)
plt.show()
```

## Make all the values in all the series that are less than zero equal to zero

```python
sins_copy = sins.copy()
sins_copy[sins_copy < 0] = 0
for s in sins_copy:
    plt.plot(s)
plt.show()
    
```

## Now make all the values that go above the second line (the one you moved up 2) equal the values on that second line
- In other words, no line should go above the one you moved up, but follow it until it drops below it again
- You can do it element wise with 2 loops
- Or use logical indexing

```python
sins_copy = sins.copy()
s_check = sins[1] + 2

for i, s in enumerate(sins_copy):
    for idx, (value, check) in enumerate(zip(s, s_check)):
        if value > check:
            s[idx] = check
    if i == 1:
        plt.plot(s + 2)
    else:
        plt.plot(s)
```

```python
sins_copy = sins.copy()
s_check = sins[1] + 2

check_array = sins_copy > s_check
for i, (check, s) in enumerate(zip(check_array,sins_copy)):
    for idx, (chk, value, new_value) in enumerate(zip(check, s, s_check)):
        if chk:
            s[idx] = new_value
    if i == 1:
        plt.plot(s + 2)
    else:
        plt.plot(s)
```
